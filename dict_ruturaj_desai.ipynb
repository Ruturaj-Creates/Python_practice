```python
info={"name":"ruturaj","age":23,"city":"pune"}

info["city"]="mumbai"

print(info)
```

5. Create a dictionary and remove one key using pop().


```python
emp={"id":101,"name":"viren"}

emp.pop("name")

print(emp)
```

6. Create a dictionary and remove the last inserted item using popitem().


```python
data={"id":101,"name":"vaishnav","salary":50000}

data.popitem()

print(data)
```
