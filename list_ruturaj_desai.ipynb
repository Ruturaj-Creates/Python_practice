# Basic List Operations 

1. Create a list of 5 numbers and add a new number to the end.


```python
lst1=[15,25,35,45,55]

lst1.append(65)

print(lst1)
```

2. Make a list of fruits and insert a fruit at the second position.


```python
fruits=["apple","kiwi","banana","watermellon"]

fruits.insert(1,"Mango")

print(fruits)
```

3. Create a list of names and remove one name from the list.


```python
name=["aryan",'rahul','nitin','viren']

name.remove("nitin")

print(name)
```

4. Create a list of 4 items and change the value at index 2.


```python
lst2=['books','laptop','pens','marker']

lst2[2]='notebook'                       # to replace values ---> list[index]=new value

print(lst2)
```

5. Make a list of numbers and print only the first three elements.


```python
nums=[3,6,9,12,15,18]

print(nums[:3])
```

6. Create a list of colors and find the index of a specific color.


```python
colors=['orange','white','green','yellow']

colors.index('yellow')
```

7. Create a list of 10 numbers and print how many times “5” appears.


```python
numbers=[2,4,5,7,8,10,5,1,5,12]

numbers.count(5)
```

8. Make a list of 6 items and delete the item at index 3.


```python
lst2=['books','laptop','pens','marker','notebook','board']

lst2.pop(3)

print(lst2)
```

9. Create a list of animals and add three more animals using extend().


```python
animals=['lion','tiger','wolf']

animals.extend(['rabbit','deer','elephant'])

print(animals)
```

10. Create a list and remove the last item using pop().


```python
lst2=['books','laptop','pens','marker','notebook','board']

lst2.pop()

print(lst2)
```

11. Create a list of prices and sort them in ascending order.


```python
price=[500,900,100,600,200]

price.sort()

print(price)
```

12. Make a list of countries and reverse the order of the list.


```python
country=['germany','england','brazil','india']

country.reverse()                 #country[::-1]
print(country)
```

13. Create a list of words and check if “apple” is in the list.


```python
fruits=["apple","kiwi","banana","watermellon"]

"apple" in fruits
```

14. Create a list of numbers and print the largest and smallest number.


```python
price=[500,900,100,600,200]
print(max(price))
print(min(price))
```

15. Make a list of 5 items and print each item using a loop.


```python
fruits=["apple","kiwi","banana","watermellon",'peach']

for i in fruits:
    print(i)
```

16. Create two lists and combine them into one list.


```python
fruits=["apple","kiwi","banana","watermellon",'peach']

animals=['lion', 'tiger', 'wolf', 'rabbit', 'deer', 'elephant']

fruits.extend(animals)   # fruits= fruits + animals

print(fruits)
```

17. Make a list and replace the last item with a new value.


```python
animals=['lion', 'tiger', 'wolf', 'rabbit', 'deer', 'elephant']

animals[-1]="shark"

print(animals)
```

18. Create a list of numbers and print only the even numbers.


```python
num=[11,12,13,14,15,16,17,18]

for i in num:
    if i%2==0:
        print(i)
```

19. Create a list and clear all items from the list.


```python
prices=[500,900,100,600,200,300]

prices.clear()

print(prices)
```

```python
list1=[1,2,3,4,5]

list2= list1.copy()

print(list2)

```
