## **1. Print every element inside this nested list:**

## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here
data=[[10,20], [30], [5, 7, 8]]
sum=0
for item in data:
    for num in item:
        sum+=num
print(sum)
````

## **8. Flatten a nested list into a single list.**
`[[1,2], [3,4], [5]] â†’ [1,2,3,4,5]`

````python
# Write your answer here
numbers=[[1,2], [3,4], [5]] 
lst=[]
for num in numbers:
    for i in num:
        lst.append(i)

print(lst)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here
numbers = [[12, -45, 78],[-34, 90, 23],[56, 18, 99],[11, 67, -42]]
for num in range(len(numbers)):
    for i in range(len(numbers[num])):
        if (numbers[num][i])<0:
            (numbers[num][i])=0
print(numbers)
````

## **20. Convert all inner dictionary keys to uppercase.**

````python
# Write your answer here
students = {
    "Aarav": {"math": 80, "science": 78},
    "Meera": {"math": 91, "science": 92},
    "Rohan": {"math": 85, "science": 85},
    "Ananya": {"math": 88, "science": 88},                      #revise
    "Kabir": {"math": 90, "science": 95},
    "Siya": {"math": 82, "science": 81}
}

new_data = {}
for i, j in students.items():
    new_data[i] = {}
    for key, value in j.items():
        new_data[i][key.upper()] = value

print(new_data)


    
````
